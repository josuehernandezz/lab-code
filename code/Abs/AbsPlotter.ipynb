```python
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd

for i in range(2,7):

    samp_name = 'OA ' + str(i-1)
    save_name = samp_name + 'Abs_pt'

    #File paths of absorbance and PL data
    abs_path = '/Users/josuehernandez/jh_sheldon_group/data/Abs/Ligand Absorbances/OA Standard Solutions/OA_50_' + str(i) + '000_JH_4-5-23_Absorbance__0__1.txt'
    #abs_path = '/Users/josuehernandez/jh_sheldon_group/data/Abs/Ligand Absorbances/OAm Standard Solutions/OAm_50_' + str(i) + '000_JH_4-5-23_Absorbance__0__1.txt'
    # nanocrystal plot pl_path = ''

    # Figure Save Location
    save_path = '/Users/josuehernandez/jh_sheldon_group/figures/Abs_PL_figures/Ligands/OA/' + save_name + '.png'
    #save_path = '/Users/josuehernandez/jh_sheldon_group/figures/Abs_PL_figures/Ligands/OAm/' + save_name + '.png'

    abs_data = pd.read_csv(abs_path, delimiter='\t', names=('wavelength', 'intensity'))
    # nanocrystal plot pl_data = pd.read_csv(pl_path, delimiter='\t', names=('wavelength', 'intensity'))

    # Figure and axes titles
    title = 'Oleic Acid'
    x_label = 'Wavelength (nm)'
    y_label = 'Normalized Intensity (a.u.)'

    # Abs Data
    wavelength = abs_data.wavelength
    f_int = abs_data.intensity
    abs_norm_inten = f_int / np.max(f_int)

    newWavelength = wavelength[57:700]

    # # PL data
    # pl_wavelength = pl_data.wavelength
    # pl_inten = pl_data.intensity
    # pl_norm_inten = pl_inten / np.max(pl_inten)

    # Format Flots
    float_formatter = "{:.1f}".format

    # Plots
    plt.plot(wavelength, f_int, label='Abs ' + samp_name)
    plt.plot(newWavelength[f_int[57:700].argmax(axis=0)+57],max(f_int[57:700]), 
             marker='o', 
            label='Max '
            + str(float_formatter(newWavelength[f_int[57:700].argmax(axis=0)+57])) + ' nm, ' 
            + str(float_formatter(f_int[f_int[57:700].argmax(axis=0)+57])), markerfacecolor='black')

    plt.xlim([200, 400])
    plt.ylim([0, 1])
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.title(title)
    plt.legend()
    plt.savefig(save_path, dpi=1200)
plt.show()
```
